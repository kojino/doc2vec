```python
# Import libraries
import gensim.models as g
from collections import namedtuple
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sys 
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity
import scipy as sp

#parameters
doc2vec_dir="model/enwiki_dbow/doc2vec.bin"
```

```python
# Read in complaints
comparison = pd.read_csv("data/don_quixote_comparsion.csv",encoding='latin-1')
comparison.shape
```

```python
comparison
```

```python
comparison = comparison.dropna()
```

```python
model = g.Doc2Vec.load(doc2vec_dir)
```

```python
doc_vec = comparison.applymap(model.infer_vector)
```

```python
v1_vec = pd.DataFrame(doc_vec['Version 1'].tolist())
v2_vec = pd.DataFrame(doc_vec['Version 2'].tolist())
```

```python
dist = pd.DataFrame(sp.spatial.distance.cdist(v1_vec,v2_vec,metric='cosine'))
```

```python
dist.nsmallest(31,9)
```

```python
def pos_of_equiv(col):
    return np.where(col.nsmallest(col.shape[0]).index==col.name)[0][0]
```

```python
np.sum(dist.apply(pos_of_equiv,axis=0) < 10)
```
